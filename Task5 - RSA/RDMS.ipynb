```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
import matplotlib.pyplot as plt

def plot_rdm_from_file(file_path):
    # Load feature matrix from .npy file
    feature_matrix = np.load(file_path, allow_pickle=True)
    
    # Reshape the feature matrix to remove the singleton dimension
    feature_matrix = np.squeeze(feature_matrix)
    
    # Compute pairwise dissimilarities
    dissimilarities = pdist(feature_matrix, metric='euclidean')
    
    # Construct RDM
    rdm = squareform(dissimilarities)
    
    # Plot RDM
    plt.imshow(rdm, cmap='viridis', interpolation='nearest', vmin=np.min(dissimilarities), vmax=np.max(dissimilarities))
    plt.colorbar()
    plt.title('Representational Dissimilarity Matrix')
    plt.xlabel('Video Index')
    plt.ylabel('Video Index')
    plt.show()

# Test the function with a file path
layer_6 = plot_rdm_from_file(r'C:\Users\andre\OneDrive\Documents\GitHub\algonuts_2021_andrea\PCA\PCA_slowfast\train_block_6.npy')
layer_6
```
